```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("uber.csv")
```

```python
df.head()
```

```python
df.dropna(inplace = True)
df.drop(["Unnamed: 0"], axis = 1, inplace = True)
```

```python
def hav(lat1, lat2, lon1, lon2):
    lat1, lat2, lon1, lon2 = map(np.radians, [lat1, lat2, lon1, lon2])
    diff_lat = lat2-lat1
    diff_long = lon2 - lon1
    return 4000*(np.arccos(np.sin(lat1)*np.sin(lat2) + np.cos(lat1)*np.cos(lat2)*np.cos(diff_long)))
```

```python
df["distance"] = hav(df["pickup_latitude"], df["dropoff_latitude"], df["pickup_longitude"], df["dropoff_longitude"])
```

```python
df.head()
```

```python
df["hours"] = df["pickup_datetime"].str.slice(11, 13)
```

```python
df["hours"] = df["hours"].astype(int)
```

```python
df["month"] = df["pickup_datetime"].str.slice(5, 7).astype(int)
```

```python
df.head()
```

```python
df.drop(df[df["fare_amount"] <= 0].index, inplace = True)
```

```python
def find_latitude_outliers(series):
    ans = []
    for i in series:
        if(i<-90 or i>90):
            ans.append(i)
    return ans

def find_longitude_outliers(series):
    ans = []
    for i in series:
        if(i<-180 or i>180):
            ans.append(i)
    return ans


pickup_longitude_outliers = find_longitude_outliers(df['pickup_longitude'])
pickup_latitude_outliers = find_latitude_outliers(df['pickup_latitude'])
dropoff_longitude_outliers = find_longitude_outliers(df['dropoff_longitude'])
dropoff_latitude_outliers = find_latitude_outliers(df['dropoff_latitude'])

print(len(pickup_longitude_outliers))
print(len(pickup_latitude_outliers))
print(len(dropoff_longitude_outliers))
print(len(dropoff_latitude_outliers))
```

```python
df.describe()
```

```python
from matplotlib.cbook import boxplot_stats
pass_outliers = boxplot_stats(df["passenger_count"])[0]["fliers"]
import seaborn as sns
sns.boxplot(df["passenger_count"])

filter = df['passenger_count']<=6
df = df.loc[filter]
```

```python
# Outliers
Q1 = df["fare_amount"].quantile(0.25)
Q3 = df["fare_amount"].quantile(0.75)

IQR = Q3 - Q1
df[df["fare_amount"] < (Q1 - IQR*1.5)]["fare_amount"] = Q1
df[df["fare_amount"] > (Q3 + IQR*1.5)]["fare_amount"] = Q3
```

```python
df = df[(df["distance"] > 0) & (df["distance"] < 50)]
```

```python
df.drop(["pickup_longitude", "dropoff_longitude", "pickup_latitude", "dropoff_latitude"], axis = 1, inplace = True)
```

```python
def gradient_descent(start, gradient, learning_rate, iterations, stop_val = 0.1):
    history = [start]
    value = start
    for i in range(iterations):
        diff_value = gradient(value)*learning_rate
        
        if(diff_value <= stop_val):
            break
        
        value -= diff_value
        
        history.append(value)
        
    return history, value
```

```python
def function(x):
    return x**2 + 2*x - 4

def gradient(x):
    return 2*x + 2
```

```python
print(gradient_descent(2, gradient, 0.05, 1000))
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
df.dropna(inplace = True)
```

```python
from matplotlib.cbook import boxplot_stats
#Outliers for Age 
distance_outliers = boxplot_stats(df.Age)[0]['fliers']
distance_outliers.sort()
sns.boxplot(df['Age'])
print(distance_outliers)
print(len(distance_outliers))
# Get rid of outliers here, flooring and capping
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15)
```

```python
from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors = 2)

neigh.fit(X_train, y_train)
```

```python
neigh.score(X_test, y_test)
```

```python
y_pred = neigh.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score

print("Confusion Matrix", confusion_matrix(y_test, y_pred), sep="\n")

print("Accuracy", accuracy_score(y_test, y_pred))
print("Precision", precision_score(y_test, y_pred))
print("Recall", recall_score(y_test, y_pred))
print("Error", 1-accuracy_score(y_test, y_pred))
```
